```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n_customers= 10
p_success = 0.9 #claimed null hypothesis

ideal_dist = stats.binom(n_customers , p_success)
x_values = np.arange(n_customers + 1)
y_values= ideal_dist.pmf(x_values)
plt.stem(y_values)
```
