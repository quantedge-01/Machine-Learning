```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
x , y = datasets.make_blobs(n_samples = 200 , centers= 2 , n_features= 2 , center_box=(0,10))
plt.plot(x[: , 0][y == 0 ]  , x[: , 1][y==0], 'g^') # data point for one class 
plt.plot(x[: , 0][y == 1 ]  , x[: , 1][y==1], 'bs') # data point of amother class
plt.show()
```

```python
''' 
ax + by + c = 0 # general equation of a line
slope = -a/b
intercep = -c/b
w = [1,2] # random assignment of values
w.dot(x) =  1(x1)  + 2(x2) + w0
'''
```

```python
def show(w, w0, x , y): # w represents w1w2.. of ( w1x1 + w2x2 + ... + wnxn + w0 = 0) equation
    slope = -w[0]/w[1] #-a/b
    intercept = -w0/w[1] #-c/b
    x_hplane = np.arange(0,10)
    y_hplane = slope * x_hplane + intercept # y_intercept --> y = mx+c
    plt.plot(x_hplane , y_hplane)
    plt.plot(x[: , 0][y == 1 ]  , x[: , 1][y==1], 'g^')
    plt.plot(x[: , 0][y == -1 ]  , x[: , 1][y==-1], 'bs')
    # plt.show()
```

```python
w = np.random.normal(size = (2,)) # size only 2 and takinng normal distribution
w0 = np.random.normal() #
```

```python
show(w, w0, x , y) # x --> features , y --> classes
```

```python
def classifier(x,y,w,w0):
    for i in range(5000):
        miss_classified_points = 0
        for point in range(len(x)):
            half_space = np.sign(np.dot(w,x[point]) + w0)
            if half_space != np.sign(y[point]):
                miss_classified_points+= 1
                w = w + x[point]*np.sign(y[point]) # weight vector that defines decision boundary
                w0 = w0 + np.sign(y[point]) # bias or intercept that shifts the boundary
                break

        if miss_classified_points == 0: #from here we can control the accuracy level
            show(w,w0, x,y)
            # print(i)
            print('Decision boundary marked')
            break

        if i % 5 == 0: 
            show(w,w0, x,y)

    return w,w0
```

```python
classifier(x,y,w,w0)
```
